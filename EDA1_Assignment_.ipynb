```python
import pandas as pd
df=pd.read_csv("/content/Cardiotocographic.csv")
df
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handle missing values by imputing with mean
df.fillna(df.mean(), inplace=True)
```

```python
print(df.dtypes)
```

```python
# Detect and treat outliers (e.g., using Z-score method)
import numpy as np
from scipy import stats
z_scores = np.abs(stats.zscore(df))
print(np.where(z_scores > 3))
```

```python
df.describe()
```

```python
# Highlight interesting findings
print("The mean value of LB is around 134, which is slightly higher than the median value of 132.")
print("The standard deviation of AC is relatively high, indicating a wide range of values.")
```

```python
# Histograms for numerical variables
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['LB'], kde=True)
plt.title('Distribution of LB')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['AC'],kde=True)
plt.title('Distribution of AC')
plt.show()
```

```python
# Scatter plot for relationships between variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LB', y='AC', data=df)
plt.title('Relationship between LB and AC')
plt.show()
```

```python
# Correlation heatmap
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Identify correlations between variables
print("The correlation between LB and AC is around 0.7, indicating a strong positive relationship.")
print("The correlation between UC and DL is around -0.5, indicating a moderate negative relationship.")
```
